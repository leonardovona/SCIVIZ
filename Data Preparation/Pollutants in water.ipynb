```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./F2_1_Total Releases at National Level into Water.csv', sep=',')
```

```python
df = df.dropna(how="all")
```

```python
countries = df['countryName'].unique()
countries
```

```python
countries = df['countryName'].unique()
IDs = ["AT", "BE", "BG", "HR", "CY", "CZ", "DK", "EE", "FI", "FR", "DE", "EL",
       "HU", "IE", "IT", "LV", "LT", "LU", "MT", "NL", "PL", "PT", "RO", "SK",
       "SI", "ES", "SE"]

countries_to_IDs = dict(zip(countries, IDs))
df['countryName'] = df['countryName'].map(countries_to_IDs)
df['countryName']
```

```python
df['reportingYear'] = df['reportingYear'].astype(int)
```

```python
df = df.pivot_table('emissions', ['reportingYear', 'countryName'], 'pollutant')
```

```python
heavy_metals = ['Cadmium and compounds (as Cd)', 'Mercury and compounds (as Hg)', 'Nickel and compounds (as Ni)', 'Lead and compounds (as Pb)']
heavy_metals_series = df[heavy_metals].sum(axis=1)
```

```python
df = df[['Total nitrogen', 'Total phosphorus', 'Total organic carbon(as total C or COD/3) (TOC)']]
df['Heavy metals (Cd, Hg, Ni, Pb)'] = heavy_metals_series
df.rename({'Total nitrogen': 'Nitrogen', 'Total phosphorus': 'Phosphorus', 'Total organic carbon(as total C or COD/3) (TOC)': 'TOC'}, axis=1, inplace=True)
```

```python
df
```

```python
import itertools

index_pairs = df.index.to_list()

for (year, ID) in itertools.product(range(2007,2020), IDs):
    if not((year, ID) in index_pairs):
      print((year, ID))
      df.loc[(year, ID), :] = (np.NaN, np.NaN, np.NaN, np.NaN)
```

```python
newidx = [df.index.levels[0],
          df.index.levels[1].append(pd.Index(data=['EU']))]

df = df.reindex(pd.MultiIndex.from_product(newidx, names=df.index.names))
```

```python
for year in range(2007, 2020):
  df.loc[(year, "EU"), :] = df.loc[year].sum()
```

```python
df = df.sort_index()
```

```python
df['Total'] = df.sum(axis=1)
```

```python
df.replace(np.NaN, 0, inplace=True)
```

```python
df.loc[2007]
```

```python
df = df.reset_index(level=[0,1])
```

```python
df = df[(df['reportingYear'] >= 2007) & (df['reportingYear'] <= 2019)]
```

```python
df.to_csv(path_or_buf="./water_pollutants.csv")
```
