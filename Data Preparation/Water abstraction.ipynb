```python
import pandas as pd
import numpy as np
```

Import raw data.

```python
df = pd.read_csv('./water-abstraction-by-economic-sector-1.csv', sep=',')
```

```python
df
```

```python
df.info()
```

Remove extra empty rows.

```python
df = df.dropna(how="all")
```

Rename columns.

```python
df.rename({'Country:text': 'country', 'Economic activity:text': 'economic_activity', 'Year:text': 'year', 'Value:number': 'value'}, axis=1, inplace=True)
```

```python
df
```

Remove non-EU countries.

```python
non_EU_countries = ['Albania', 'Bosnia and Herzegovina', 'Iceland', 'Kosovo*', 'Montenegro', 'North Macedonia', 'Norway', 'Serbia', 'Switzerland', 'Türkiye']

df = df[~df.country.isin(non_EU_countries)]
```

Replace country names with their IDs.

```python
countries = sorted(df['country'].unique())
IDs = ["AT", "BE", "BG", "HR", "CY", "CZ", "DK", "EE", "FI", "FR", "DE", "EL",
       "HU", "IE", "IT", "LV", "LT", "LU", "MT", "NL", "PL", "PT", "RO", "SK",
       "SI", "ES", "SE"]

countries_to_IDs = dict(zip(countries, IDs))
df['country'] = df['country'].map(countries_to_IDs)
df['country']
```

```python
df[df['year'] == 2007]
```

```python
df.economic_activity.unique()
```

```python
df
```

Explode economic activity column values into distinct columns.

```python
df = df.pivot_table(index=['year', 'country'], columns='economic_activity', values='value')
```

```python
df
```

Reindex dataframe to add values for EU.

```python
newidx = [df.index.levels[0],
          df.index.levels[1].append(pd.Index(data=['EU']))]

df = df.reindex(pd.MultiIndex.from_product(newidx, names=df.index.names))
```

Replace NaN values with 0.

```python
df.replace(np.NaN, 0, inplace=True)
```

Insert EU values as sum of the other countries' values.

```python
for year in range(2007, 2020):
  df.loc[(year, "EU"), :] = df.loc[year].sum()
```

Sort the data frame by year and country.

```python
df = df.sort_index()
```

Replace possible NaN values with 0.

```python
df.replace(np.NaN, 0, inplace=True)
```

```python
df.loc[2007]
```

Restore single index.

```python
df = df.reset_index(level=[0,1])
```

```python
df[df.country == "EU"]
```

Filter years from 2007 to 2019.

```python
df = df[(df['year'] >= 2007) & (df['year'] <= 2019)]
```

Export dataset.

```python
df.to_csv(path_or_buf="./water_abstraction.csv")
```
